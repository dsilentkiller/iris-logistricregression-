```python
def single_features_prediction(petal_length):
    '''predicts the iris species given the petal length.'''
    
    if petal_length<2.5:
        return 0
    elif petal_length <4.8:
        return 1
    else:
        return 2
        
```

```python
def plot_incorrect_predictions(df_predictions, x_axis_feature, y_axis_feature):
    fig, axs = plt.subplots(2,2,figsize=(10,10))
    axs=axs.flatten()
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue ="target", data =df_predictions,ax=axs[0])
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue ="target", data =df_predictions,ax=axs[1])
    plt.show()
```
